```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tushare as ts

with open('token.txt', 'r') as token_file:
    token = token_file.read().replace('\n', '')
ts.set_token(token)
pro = ts.pro_api()

sz50_bs = pd.read_csv('csv\sz50_debt.csv', encoding='utf-8-sig')
st_bs = pd.read_csv('csv\st_debt.csv', encoding='utf-8-sig')
db = pro.daily_basic(ts_code='', 
                    trade_date='20171229', 
                    fields='ts_code,trade_date,close,total_share,float_share,total_mv,circ_mv')

sz50_bs = pd.merge(sz50_bs, db, how='left', on='ts_code')
st_bs = pd.merge(st_bs, db, how='left', on='ts_code')
```

```python
sz50_bs = sz50_bs.drop(['Unnamed: 0'], axis=1) .dropna()
st_bs = st_bs.drop(['Unnamed: 0'], axis=1) .dropna()
```

```python
sz50_bs.to_csv('csv\sz50_bs.csv', encoding='utf-8-sig')
st_bs.to_csv('csv\st_bs.csv', encoding='utf-8-sig')
```
